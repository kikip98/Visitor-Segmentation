```python
#Import libraries
import pandas as pd
from google.cloud import bigquery
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', -1)
```

```python
#Query data from fact_state_info table
client = bigquery.Client()


sql = """
SELECT ip_address, interaction_id, ts, state_id, page_url_clean, page_referrer, cart,campaign_id, discount_code, fingerprint
FROM `cloudiq-jay-prod-01.analytics_platform.fact_state_info`
WHERE tag_id = '9b2bea23-5e43-48fa-b615-f44ebacb02aa-jjztnjrv' 
AND DATE(ts) BETWEEN "2022-03-01" AND "2022-03-31"
ORDER BY ip_address, interaction_id, ts
"""

fact_state_info = client.query(sql).to_dataframe()
fact_state_info.head(3)
```

```python
#Save data to csv
fact_state_info.to_csv("fact_state_info.csv")
```

```python
fact_state_info['interaction_id'].nunique()
```

```python
#Query data from fact_interaction_info table

sql = """
SELECT *
FROM `cloudiq-jay-prod-01.analytics_platform.fact_interaction_info`
WHERE tag_id = '9b2bea23-5e43-48fa-b615-f44ebacb02aa-jjztnjrv' 
AND DATE(start_ts) BETWEEN "2022-03-01" AND "2022-03-31";
"""

fact_interaction_info = client.query(sql).to_dataframe()
fact_interaction_info.head()
```
